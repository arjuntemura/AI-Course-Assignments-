```python
#importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#read data
data=pd.read_csv("roo_data.csv")
data.head(5)
```

```python
print("The dimensions of the data is " + str(data.shape))
```

```python
data.info() #check for any missing fields 
```

```python
#remove unnecessary data
data.drop(['Extra-courses did', 'reading and writing skills','Taken inputs from seniors or elders','interested in games','Interested Type of Books','Salary Range Expected','In a Realtionship?','Gentle or Tuff behaviour?','Salary/work','Introvert','hackathons','Job/Higher Studies?','hard/smart worker', 'Hours working per day','talenttests taken?', 'olympiads','can work long time before system?'], axis=1, inplace=True)
data.shape #we have 22 columns now
```

```python
#replace 'yes'/'no' with 1/0
data["self-learning capability?"].replace(to_replace=['no', 'yes'], value=[0, 1], inplace=True)
data['self-learning capability?'].head()
```

```python
#binary encoding 
technical=pd.get_dummies(data['Management or Technical'], drop_first=True)
technical.head()
```

```python
#adding this field to main dataframe
data=pd.concat([data, technical], axis=1)
data.drop(['Management or Technical'], axis=1, inplace=True)
data.head()
```

```python
#binary encoding
subjects=pd.get_dummies(data['Interested subjects'], drop_first=True)
subjects.head()
```

```python
#binary encoding
career=pd.get_dummies(data['interested career area '], drop_first=True)
career.head()
```

```python
#binary encoding 
company=pd.get_dummies(data['Type of company want to settle in?'], drop_first=True)
company.head()
```

```python
#binary encoding
certificates=pd.get_dummies(data['certifications'], drop_first=True)
certificates.head()
```

```python
#adding these fields to dataframe
data=pd.concat([data, subjects, career, certificates], axis=1)
data.drop(['Type of company want to settle in?', 'interested career area ','Interested subjects','certifications'], axis=1, inplace=True)
data.head()
```

```python
#new mean of grades field
df=data.iloc[:,:8]
df['Computer Science Percentage'] = df.mean(axis=1)
df
```

```python
#add the mean grade field to the dataframe
data.drop(['Acedamic percentage in Operating Systems',	'percentage in Algorithms'	,'Percentage in Programming Concepts','Percentage in Software Engineering',	'Percentage in Computer Networks',	'Percentage in Electronics Subjects'	,'Percentage in Computer Architecture',	'Percentage in Mathematics'],axis=1, inplace=True)
data=pd.concat([data,df['Computer Science Percentage']], axis=1)
#removing attributes based on hit and trial on accuracy
data.drop(['workshops','worked in teams ever?','memory capability score','public speaking points'], inplace=True, axis=1)
```

```python
#unique values of Jog Roles
no_of_categories=data['Suggested Job Role'].unique()
no_of_categories
```

```python
#merging job roles in broad categories 
data['Suggested Job Role'].replace(to_replace=['Database Developer', 'Portal Administrator',
       'Systems Security Administrator', 'Business Systems Analyst',
       'Software Systems Engineer', 'Business Intelligence Analyst',
       'CRM Technical Developer', 'Mobile Applications Developer',
       'UX Designer', 'Quality Assurance Associate', 'Web Developer',
       'Information Security Analyst', 'CRM Business Analyst',
       'Technical Support', 'Project Manager',
       'Information Technology Manager', 'Programmer Analyst',
       'Design & UX', 'Solutions Architect', 'Systems Analyst',
       'Network Security Administrator', 'Data Architect',
       'Software Developer', 'E-Commerce Analyst',
       'Technical Services/Help Desk/Tech Support',
       'Information Technology Auditor', 'Database Manager',
       'Applications Developer', 'Database Administrator',
       'Network Engineer', 'Software Engineer', 'Technical Engineer',
       'Network Security Engineer',
       'Software Quality Assurance (QA) / Testing'], 
       value=['Data Analyst', 'Technical Analyst','Security/Networking Analyst','Business Analyst','Software/UX Analyst','Business Analyst',
             'Technical Analyst','Software/UX Analyst',
             'Software/UX Analyst', 'Technical Analyst','Software/UX Analyst',
             'Security/Networking Analyst','Business Analyst','Technical Analyst','Business Analyst',
             'Security/Networking Analyst', 'Software/UX Analyst',
             'Software/UX Analyst', 'Technical Analyst', 'Technical Analyst','Security/Networking Analyst', 'Data Analyst', 
             'Software/UX Analyst', 'Business Analyst', 'Technical Analyst','Security/Networking Analyst','Data Analyst','Software/UX Analyst',
             'Data Analyst', 'Security/Networking Analyst', 'Software/UX Analyst','Technical Analyst','Security/Networking Analyst','Technical Analyst'], inplace=True)
```

```python


#New categories of Job roles
data['Suggested Job Role'].unique()
```

```python

data.head()
```

```python

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.manifold import TSNE
from sklearn.model_selection import KFold

X=data.drop("Suggested Job Role", axis=1)
y=data["Suggested Job Role"]

#split training and testing data
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.1, random_state=1)

#Scaling data 
scaler=StandardScaler()
scaler.fit(X_train)
X_train=scaler.transform(X_train)
X_test=scaler.transform(X_test)


print("Normalized data: ")
X_train
```

```python

#ANN model 
from sklearn.neural_network import MLPClassifier
MLP=MLPClassifier(hidden_layer_sizes=(40,20,2), max_iter=1000,activation = 'relu',solver='adam',random_state=1)
MLP.fit(X_train, y_train.values.ravel())
```

```python
predictions=MLP.predict(X_test)
predictions
```

```python
#classwise accuracy
print("Classification report: ")
print(classification_report(y_test, predictions))
```

```python
print("Confusion matrix: ")
print(confusion_matrix(y_test, predictions))
from sklearn.metrics import accuracy_score
acc = accuracy_score(predictions , y_test)
print("The accuracy of model is", acc*100,"%")
```
